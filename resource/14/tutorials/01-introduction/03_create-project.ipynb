# Create your first project

1. **Create project directory**
    ```
    mkdir onlineshop
    ```
2. **cd into the created directory**
    ```
    cd onlineshop
    ```
3. **Install django in venv**
    ```
    pipenv install django
    ```
    - copy Virtualenv location: `/home/$USER/.local/share/virtualenvs/onlineshop-hjMltqKS`
4. **Open vscode**
    ```
    code .
    ```
5. **Checkout *Pipfile* and *Pipfile.lock***
6. **Activate virtual env**
    ```
    pipenv shell
    ```
7. **Django admin commands**
    
    - **django-admin commands**
        ```
        django-admin
        ```
        - *check*
        - *compilemessages*
        - *createcachetable*
        - *dbshell*
        - *diffsettings*
        - *dumpdata*
        - *flush*
        - *inspectdb*
        - *loaddata*
        - *makemessages*
        - *makemigrations*
        - *migrate*
        - *runserver*
        - *sendtestemail*
        - *shell*
        - *showmigrations*
        - *sqlflush*
        - *sqlmigrate*
        - *sqlsequencereset*
        - *squashmigrations*
        - *startapp*
        - *startproject*
        - *test*
        - *testserver*

    - **create django project**
        ```
        django-admin startproject onlineshop # create project with a redundant directory
        django-admin startproject onlineshop . # create project without redundancy
        ``` 
8. **runserver**
    ```
    django-admin runserver # raise exception due to unconfigured project
    python manage.py runserver 9000 # default port: 8000
    ```
9.  **Checkout browser** *`http://127.0.0.1:9000/`*

## Other way

1. Open up **CMD**
2. Create a diretory wherever you want ```mkdir PROJECT_NAME```
3. Go to the directory ```cd PROJECT_NAME``` 
4. Install *virtualenv*, ```pip install virtualenv```
5. Create a virtualenv, 
    ```
    python -m virtualenv YOUR_VENV_NAME  # like python -m virtualenv venv 
    python3 -m virtualenv YOUR_VENV_NAME  # python -m virtualenv venv
    ```
6. Activate virtualenv
    ```
    .\YOUR_VENV_NAME\Scripts\activate  # like  .\venv\Scripts\activate
    ```
7. Install django 
    ```
    pip install django
    ```
8. Use *django-admin* commands to create a django project.
    ```
    django-admin startproject onlinestore . 
    ```
9. Run server
    ```
    python manage.py runserver 
    ```
1.  Check out website at *`http://127.0.0.1:8000/`*
